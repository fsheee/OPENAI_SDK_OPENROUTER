# Using the OPEN ROUTER API directly

```python
import requests
import json

response = requests.post(
  url=f"https://openrouter.ai/api/v1/chat/completions",
  headers={
    "Authorization": f"Bearer {OPENAPI_SDK_OPENROUTER}",
  },
  data=json.dumps({
    "model": MODEL,
    "messages": [
      {
        "role": "user",
        "content": "Hi,there"
      }
    ]
  })
)

print(response.json())
```

## 2.USING OPEN API AGENTS SDK

```python
import asyncio
from openai import AsyncOpenAI # chat completions
from agents import Agent, OpenAIChatCompletionsModel, Runner, set_tracing_disabled

set_tracing_disabled(disabled=True) # Open AI Tracing == Disable

client = AsyncOpenAI(
    api_key=OPENAPI_SDK_OPENROUTER,
    base_url=BASE_URL
)


async def main():
    # This agent will use the custom LLM provider
    agent = Agent(
        name="Assistant",
        instructions="You only respond in englis.",
        model=OpenAIChatCompletionsModel(model="mistralai/mistral-7b-instruct:free", openai_client=client),
    )

    result = await Runner.run(
        agent, # starting agent
        "Tell me about recursion in bussiness programming.", # request
    )
    print(result.final_output)

if __name__ == "__main__":
  asyncio.run(main())
```
